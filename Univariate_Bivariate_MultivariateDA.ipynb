```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("alg_fire_1.xlsx")
```

```python
df["Classes  "].value_counts()
```

```python
for i in range(0,len(df)):
  if df["Classes  "][i] == "not fire   " or df["Classes  "][i] == "not fire " or df["Classes  "][i] == "not fire " or df["Classes  "][i] == "not fire":
    df["Classes  "][i] = "not fire"
  else:
    df["Classes  "][i] = "fire"
```

#Univariate Analysis

Choosing Rain as my univariate parameter 

```python
df_not_fire = df.loc[df["Classes  "]=="not fire   "]
x = df.loc[df["Classes  "]=="not fire"]
y = df.loc[df["Classes  "]=="not fire "]
df_not_fire = df_not_fire.append(x)
df_not_fire =  df_not_fire.append(y)
```

```python
df_fire = df.loc[df["Classes  "]=="fire   "]
x = df.loc[df["Classes  "]=="fire"]
y = df.loc[df["Classes  "]=="fire "]
df_fire = df_fire.append(x)
df_fire =  df_fire.append(y)
```

```python
df_not_fire.shape
```

```python
df_fire.shape
```

###Rain

```python
plt.plot(df_not_fire["Rain "],np.zeros_like(df_not_fire))
plt.plot(df_fire["Rain "],np.zeros_like(df_fire))
plt.xlabel("Rain")
plt.show()
```

###Temperature

```python
plt.plot(df_not_fire["Temperature"],np.zeros_like(df_not_fire),'o')
plt.plot(df_fire["Temperature"],np.zeros_like(df_fire),'o')
plt.xlabel("Temperature")
plt.show()
```

###BUI

```python
plt.plot(df_not_fire["BUI"],np.zeros_like(df_not_fire))
plt.plot(df_fire["BUI"],np.zeros_like(df_fire))
plt.xlabel("BUI")
plt.show()
```

#Bivariate Analysis

```python
sns.FacetGrid(df,hue="Classes  ",size = 6).map(plt.scatter,"Rain ","Temperature").add_legend()
plt.show()
```
