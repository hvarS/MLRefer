```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A':[1,2,4,1,3,2,5,7,8],
                   'B':[2,1,1,3.5,79,0,22,1,19],
                   'C':np.zeros(9),
                   'D':np.ones(9)})
```

```python
df.head()
```

##Dropping Constant Features

From the above created dataset , it can be observed that Column C and D will not contribute highly in the case features as they are constant 

**We will use variance threshold feature given by sklearn which will remove all the columns having variance less than a given threshold**

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
vt = VarianceThreshold()
```

```python
vt.fit(df)
```

```python
vt.get_support()
```

```python
constant_cols = [column for column in df.columns if column not in df.columns[vt.get_support()]]
```

```python
len(constant_cols)
```

```python
df.drop(constant_cols,axis=1)
```
